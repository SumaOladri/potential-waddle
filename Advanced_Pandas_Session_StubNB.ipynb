### KNN Imputer

One commonly adopted strategy for addressing missing data is to employ a predictive model to estimate the absent values. This technique entails developing a separate model for each input variable containing missing entries.

The default value of K is set to 5. Although there is no definitive method for determining the ideal value of K, a commonly used heuristic suggests that the optimal K is often the square root of the total number of samples in the dataset. Typically, an odd value is chosen for K to prevent ties in decision-making. To identify the most suitable K, an error plot or accuracy plot is commonly used.


Description: KNN (K-Nearest Neighbors) imputation involves replacing missing values with the weighted average of the observed values for that variable, where the weights are determined by the similarity (closeness) of the samples in the feature space.

Procedure: For each missing value, identify the k-nearest neighbors with complete data, calculate their weighted average, and replace the missing value with this average.

Use Case: KNN imputation is effective when the missing values are not completely at random and there is a pattern to the missingness. It considers the relationships between variables and can handle complex patterns of missing data. It's suitable for both numerical and categorical variables.

###  Multivariate Imputation by Chained Equation-MICE

MICE Imputation, short for 'Multiple Imputation by Chained Equation' is an advanced missing data imputation technique that uses multiple iterations of Machine Learning model training to predict the missing values using known values from other features in the data as predictors.

How does MICE algorithm work?

Here is a quick intuition (not the exact algorithm)

1) You basically take the variable that contains missing values as a response 'Y' and other variables as predictors 'X'.

2) Build a model with rows where Y is not missing.

3) Then predict the missing observations. Do this multiple times by doing random draws of the data and taking the mean of the predictions.

Use Case:

MICE is suitable when the missing data mechanism is assumed to be Missing at Random (MAR) or Missing Completely at Random (MCAR).
It accommodates both numerical and categorical variables and captures complex relationships between variables.
MICE is particularly effective when there are interactions and dependencies between variables.

````python
df_mice = df.copy()
df_mice.head()
````

estimator: Specifies the regression model used for imputation. Different estimators may be chosen based on the characteristics of the data. BayesianRidge is a good choice for handling linear relationships with uncertainty.

n_nearest_features: Controls the number of nearest features considered during imputation. Setting it to None means all features are considered, which is suitable when the relationships are complex and involve multiple features.\

imputation_order: Defines the order in which variables are imputed. 'ascending' means imputation follows the natural order of variables. This can impact the imputation results, especially when there are dependencies between variables.

```python
def process_page(title,description):
  chunk_size = 5000
  #description = list(job_postings[job_postings['title']==title]['description'])[0]
  chunks = split_text_chunks(description,chunk_size)
  row_title = title
  data = {"Title":[],"Chunk_Text":[]}
  for idx,chunk in enumerate(chunks):
    data['Title'].append(row_title)
    data['Chunk_Text'].append(chunk)
  return pd.DataFrame(data)




def split_text_chunks(text,chunk_size):
  chunks = []
  words = text.split()
  current_chunk_word_count = 0
  current_chunk = []
  for word in words:
    if current_chunk_word_count + len(word)+1<=chunk_size:
      current_chunk_word_count += len(word)+1
      current_chunk.append(word)
    else:
      chunks.append(' '.join(current_chunk))
      current_chunk = [word]
      current_chunk_word_count = len(word)
  if current_chunk:
    chunks.append(' '.join(current_chunk))
  return chunks
```

```python
def fixed_chunk_df_creation(df):
  all_dfs = []
  for i in range(len(df)):
    title = df.loc[i]['title']
    description = df.loc[i]['description']
    chunk_df = process_page(title,description)
    if chunk_df is not None:
        all_dfs.append(chunk_df)

  fixed_chunk_df = pd.concat(all_dfs, ignore_index=True)
  return (fixed_chunk_df)
```

```python
# Define the function for calculating cosine similarity

def calculate_similarity(embedding1, embedding2):
    cosine_score = util.pytorch_cos_sim(embedding1, embedding2)
    # Convert the result to a Python float
    similarity = cosine_score.item()

    return similarity
```

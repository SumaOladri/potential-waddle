1.3 Updating Collections

```python
section_chunk_embeddings['Chunk_Text'] = section_chunk_embeddings['Chunk_Text'].apply(lambda  x:x.lower())
```

```python
section_chunk_embeddings['Metadata'] = section_chunk_embeddings.apply(lambda x : {'Title':x['Title'],'Section':x['Section_Title']},axis=1)
```

```python
section_chunk_embeddings.head()
```

```python
metadata_list = section_chunk_embeddings['Metadata'].to_list()
```

```python
collection.upsert(
                 embeddings = section_chunk_embeddings['Embeddings'].to_list(),
                 documents = section_chunk_embeddings['Chunk_Text'].to_list(),
                 metadatas = metadata_list,
                 ids = [str(i) for i in range(0,len(section_chunk_embeddings['Embeddings']))])

```

```python
#collection.get(ids = ['0','1'])
input_query = input()

results = collection.query(query_texts = input_query,
                           n_results=3,
                           where = {'Section': 'Critical response'},
                           where_document = {"$contains": "positive"},
                           include= ['documents','distances'])
```
